```python
def iqr(column):
    q1 = df[column].quantile(0.25)
    q3 = df[column].quantile(0.75)
    return q3 - q1
```

```python
summary = pd.DataFrame(columns=['Columns', 'Mean' ,'Median', 'IQR', 'Variance', 'Skewness', 'Kurtosis'])

for i in df.columns.to_list():
    summary.loc[len(summary)] = [i, round(df[i].mean(), 6), round(df[i].median(), 6), round(iqr(i), 6), round(df[i].var(), 6), round(df[i].skew(), 6), round(df[i].kurt(), 6)]
summary
```

#### Key Findings:

##### 1. Baseling Fetal Heart Rate (LB):

* The mean baseline fetal heart rate is approximately 133 beats per minute, with a median of 133 bpm.
* The interquartile range (IQR) is relatively small at 14 bpm, indicating that most observations cluster around the median.
* The variance is moderate, suggesting some variability in fetal heart rates across the dataset.
* The skewness is close to zero, indicating a relatively symmetric distribution.
* The kurtosis is negative, suggesting a slightly flatter distribution compared to a normal distribution.

##### 2. Fetal Movement (FM):

* The mean percentage of detected fetal movements is around 0.45%, with a median of 0%.
* The distriution is highly skewed to the right, indicating that most observations have no fetal movements detected, but a few have a high percentage of movements.
* The kurtosis is extremely high, indicating heavy tails and potential outliers in the distribution.

##### 3. Acceleration (AC) and Deceleration Late (DL):

* Both variables have very low means and medians, indicating that accelerations and late ddecelerations are relatively rare in the dataset.
* The skewness is positive for DL, indicating a right-skewed distribution, possibly due to a few instances of late decelerations.
* The kurtosis is higher for DL compared to AC, indicating heavier tails and potentially more extreme values.

##### 4. Uterine Contractions (UC):

* The mean and median are close, suggesting a relatively symmetric distribution.
* The skewness is close to zero, indicating a symmetric distribution.
* The kurtosis is negative, indicating a flatter distribution compared to a normal distribution.

##### 5. ASTV (Percentage of Time with Abnormal Short Term Variability) and ALTV (Percentage of Time with Abnormal Long Term Variability):

* Both variables have relatively high means and medians, indicating a significant percentage of time with abnormal variability.
* The distributions are positively skewed, indicating that most observations have lower percentages of abnormal variability, but a few have very high percentages.
