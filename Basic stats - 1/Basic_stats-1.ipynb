```python
import pandas as pd
```

```python
df = pd.read_csv('./sales_data_with_discounts.csv')
df
```

#### Descriptive Analytics for Numerical Columns

```python
df.dtypes
```

```python
numcol = df[['Volume', 'Avg Price', 'Total Sales Value', 'Discount Rate (%)', 'Discount Amount', 'Net Sales Value']]
numcol.head()
```

```python
numcol.describe()
```

```python
import numpy as np
from scipy import stats
```

```python
dicts = []
for i in numcol.columns:
    dicts.append({
    'mean' : round(np.mean(numcol[i]), 2),
    'median' : round(np.median(numcol[i]), 2),
    'mode' : stats.mode(numcol[i]).mode,
    'std' : round(np.std(numcol[i]), 2)
})
```

```python
result = pd.DataFrame({
    'Volume' : dicts[0],
    'Avg Price' : dicts[1],
    'Total Sales Value' : dicts[2],
    'Discount Rate (%)' : dicts[3],
    'Discount Amount' : dicts[4],
    'Net Sales Value' : dicts[5]
})

result
```

## Interpretations

#### Volume:
* Mean Volumn sold in 5.07 units per transaction.
* The most frequently occuring volume in a single transaction is 3 units.

#### Avg Price:

* The most common average price is 400.

#### Total Sales Value

#### Discount Rate (%)

#### Discount Amount

#### Net Sales Value
